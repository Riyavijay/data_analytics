```python
#moving average
import matplotlib.pyplot as plt
import pandas as pd
import math
r=pd.read_csv("shampoo.csv")
d=list(r['Month'])

x1=[]
for i in range(len(d)):
    x1.append(i)
y1=list(r['Sales'])
n=3
l1=[]
slic=0
for i in range(len(y1)-n):
    sum=0
    for j in range(slic,slic+n):
        sum=sum+y1[j]
    sum=sum/n
    slic=slic+1
    l1.append(sum)
print(l1)
plt.plot(l1,label="predicted")
plt.plot(y1,label="observed")
plt.legend()
plt.show()
```
